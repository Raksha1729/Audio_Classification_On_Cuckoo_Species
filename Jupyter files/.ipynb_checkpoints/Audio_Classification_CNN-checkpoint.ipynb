### Converting Waveforms to Spectrograms

```python
def get_spectrogram(waveform):
  #Converting the waveform to a spectrogram via a STFT (Short time Fourier Transform)
  spectrogram = tf.signal.stft(
      waveform, frame_length=512, frame_step=256)
  # To obtain the magnitude of the STFT
  spectrogram = tf.abs(spectrogram)
  # We also add a 'channels' dimension so spectrogram can be used as image-like
  #input data. ('batch_size', 'height', 'width', 'channels')
  spectrogram = spectrogram[..., tf.newaxis]
  return spectrogram
```

```python
def plot_spectrogram(spectrogram, ax):
  if len(spectrogram.shape) > 2:
    assert len(spectrogram.shape) == 3
    spectrogram = np.squeeze(spectrogram, axis=-1)
     # Convert the frequencies to log scale and transpose, so that the time is
    # represented on the x-axis (columns).
    # Add an epsilon to avoid taking a log of zero.
    log_spec = np.log(spectrogram.T + np.finfo(float).eps)
    height = log_spec.shape[0]
    width = log_spec.shape[1]
    X = np.linspace(0, np.size(spectrogram), num=width, dtype=int)
    Y = range(height)
    ax.pcolormesh(X, Y, log_spec)
```

## Building and Training the Model

We add Add Dataset.cache and Dataset.prefetch operations to reduce read latency while training the model

```python
train_spectrogram_ds = train_spectrogram_ds.cache().shuffle(1000).prefetch(tf.data.AUTOTUNE)
val_spectrogram_ds = val_spectrogram_ds.cache().prefetch(tf.data.AUTOTUNE)
test_spectrogram_ds = test_spectrogram_ds.cache().prefetch(tf.data.AUTOTUNE)
```

## CNN model


We use tf.Keras.Sequential  
**tf.Keras.layers.Resizing**: To downsample the input to enable the model to train faster.  
**tf.Keras.layers.Normalization** :To normalize each pixel in the image based on its mean and standard deviation.

```python
input_shape = example_spectrograms.shape[1:]
print('Input Shape:', input_shape)
num_labels = len(label_names)

#Instantiate the 'tf.keras.layers.Normalization' layer.
norm_layer = layers.Normalization()

# Fit the state of the layer to the spectrograms
# With 'Normalization.adapt'.
norm_layer.adapt(data=train_spectrogram_ds.map(map_func=lambda spec, label: spec))


model = models.Sequential([
    layers.Input(shape=input_shape),
    # Downsample the input.
    layers.Resizing(32, 32),
    # Normalize.
    norm_layer,
    layers.Conv2D(32, 3, activation='relu'),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.25),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_labels),
])

model.summary()
```

```python
from keras.utils import plot_model
from IPython.display import SVG, Image
plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)
Image('model.png',width=400, height=800)
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'],
)
```

```python
EPOCHS = 15

model_history = model.fit(
    train_spectrogram_ds,
    validation_data = val_spectrogram_ds,
    epochs =EPOCHS,
    callbacks = tf.keras.callbacks.EarlyStopping(verbose=1, patience=4),
    )
```

```python
metrics = model_history.history
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(model_history.epoch, metrics['loss'], metrics['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.ylim([0, max(plt.ylim())])
plt.xlabel('Epoch')
plt.ylabel('Loss [CrossEntropy]')

plt.subplot(1,2,2)
plt.plot(model_history.epoch, 100*np.array(metrics['accuracy']), 100*np.array(metrics['val_accuracy']))
plt.legend(['accuracy', 'val_accuracy'])
plt.ylim([0, 100])
plt.xlabel('Epoch')
plt.ylabel('Accuracy [%]')
```

```python
model.evaluate(test_spectrogram_ds, return_dict=True)
```

```python
from pydub import AudioSegment
import os

def convert_stereo_to_mono_and_resample(input_folder, base_name, target_sampling_rate=16000, target_bit_depth=16):
    # Ensure that the input folder exists
    if not os.path.exists(input_folder):
        print(f"Error: The specified folder '{input_folder}' does not exist.")
        return

    # Get a list of files in the input folder
    files = os.listdir(input_folder)

    # Filter out non-wav audio files
    wav_files = [file for file in files if file.lower().endswith(".wav")]

    # Convert stereo to mono, resample, and rename files
    for index, wav_file in enumerate(wav_files):
        input_path = os.path.join(input_folder, wav_file)
        output_name = f"{base_name}{str(index + 1).zfill(3)}.wav"
        output_path = os.path.join(input_folder, output_name)

        # Load the audio file
        audio = AudioSegment.from_file(input_path)

        # Ensure the audio is in mono
        if audio.channels > 1:
            audio = audio.set_channels(1)

        # Convert the sampling rate to 16 kHz
        audio = audio.set_frame_rate(target_sampling_rate)

        # Check the current bit depth
        current_bit_depth = audio.sample_width * 8

        # Check if the current bit depth is 32 bits
        if current_bit_depth == 32:
            # Set the target bit depth
            audio = audio.set_sample_width(target_bit_depth // 8)

        # Export the modified audio file
        audio.export(output_path, format="wav")

        print(f"Converted stereo to mono, resampled, and renamed {wav_file} to {output_name}")

if __name__ == "__main__":
    input_folder = "/content/drive/MyDrive/Colab Notebooks/Cuckoo_mono/Pacific_Koel"  # Replace with the path to your audio files
    base_name =  "pacific_koel"   # Replace with the desired base name

    convert_stereo_to_mono_and_resample(input_folder, base_name)
```

```python
from pydub import AudioSegment

def get_sequence_length(file_path):
    # Load the audio file
    audio = AudioSegment.from_file(file_path)

    # Get the duration in seconds
    duration_seconds = len(audio) / 1000  # Convert milliseconds to seconds

    # Get the sampling rate
    sampling_rate = audio.frame_rate

    # Calculate the number of samples
    num_samples = int(duration_seconds * sampling_rate)

    return num_samples

if __name__ == "__main__":
    file_path = "/content/drive/MyDrive/Colab Notebooks/XC769625-StripedCuckoo-Taperanaevia.mp3"  # Replace with the path to your audio file

    try:
        sequence_length = get_sequence_length(file_path)
        print(f"Output Sequence Length: {sequence_length} samples")
    except Exception as e:
        print(f"Error: {e}")
```

```python
from pydub import AudioSegment
import os

def convert_directory_bit_depth(directory_path, target_bit_depth=16):
    # Get a list of audio files in the directory
    audio_files = [file for file in os.listdir(directory_path) if file.lower().endswith('.wav')]

    for file_name in audio_files:
        file_path = os.path.join(directory_path, file_name)

        # Load the audio file
        audio = AudioSegment.from_file(file_path, format="wav")

        # Check the current bit depth
        current_bit_depth = audio.sample_width * 8

        # Check if the current bit depth is 32 bits
        if current_bit_depth == 32:
            # Set the target bit depth
            audio = audio.set_sample_width(target_bit_depth // 8)

            # Overwrite the original file with the converted audio
            audio.export(file_path, format="wav")
            print(f"Converted bit depth for file: {file_name}")

if __name__ == "__main__":
    directory_path = "/content/drive/MyDrive/Colab Notebooks/Cuckoo_mono/Great_otted_Cuckoo"  # Replace with the path to your audio directory

    try:
        convert_directory_bit_depth(directory_path)
        print("Conversion successful for all eligible files.")
    except Exception as e:
        print(f"Error: {e}")

```
